```python
import seaborn as sns

iris = sns.load_dataset('iris')

x = iris.iloc[:, :-1]
# 행은 모두 포함, 열은 마지막의 -1(species) 부분을 제외
x
```

```python
import random
from sklearn.impute import SimpleImputer
import numpy as np

for col in range(4): # 랜덤하게 결측값 생성
    x.iloc[[random.sample(range(len(iris)), 10)], col] = float('nan')
    
x.head()
```

```python
x.mean(axis=0) # 평균값 확인
```

```python
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
imp_mean.fit_transform(x)[0:5, :]
# NaN에 확인한 평균값을 대입
```

```python
x.median(axis=0) # 중앙값 확인
```

```python
imp_mean = SimpleImputer(missing_values=np.nan, strategy='median')
imp_mean.fit_transform(x)[0:5, :]
# NaN에 확인한 중앙값을 대입
```

```python
x.mode(axis=0) # 중앙값 확인
```

```python
imp_mean = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imp_mean.fit_transform(x)[0:5, :]
# NaN에 확인한 최빈값을 대입
```

```python
import pandas as pd

titanic_df = pd.read_csv('./titanic/train.csv')
titanic_df
```

```python
from sklearn.impute import SimpleImputer
import numpy as np

x = titanic_df[['Age']]
x
```

```python
x.mean(axis=0)
```

```python
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
imp_mean.fit_transform(x)
# NaN에 확인한 최빈값을 대입
```

```python
x1 = titanic_df[['Embarked']]
x1
```

```python
x1.mode(axis=0)
```

```python
imp_most = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imp_most.fit_transform(x1)
# NaN에 확인한 최빈값을 대입
```

```python
titanic_df[['Embarked']] = imp_most.fit_transform(x1)
titanic_df[['Age']] = imp_mean.fit_transform(x)
titanic_df
```
