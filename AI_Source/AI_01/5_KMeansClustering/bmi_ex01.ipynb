# bmi 치수 예제
- 상위 100개의 데이터를 가지고 클러스터를 만들어서 출력하기<br><br>
- `K=3`으로 처리<br><br>
- 출력의 순서<br><br>
  - `height`, `weight`, `cluster_id`<br><br>
  - 산점도 출력 (`x` = `height`, `y` = `weight`, `c` = `cluster_id`)

```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt

df = pd.read_csv('./bmi.csv')
df_1 = df.iloc[0:100, 0:2]
```

```python
info = pd.DataFrame(
  df_1,
  columns = ['height', 'weight']
)

plt.figure(figsize=(7, 5))

plt.scatter(info.values[:, 0], info.values[:, 1])

plt.title('Clustering')

plt.xlabel('height')
plt.ylabel('weight')

plt.show()
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=3, random_state=42)
km.fit(df_1)

y_pred = km.predict(df_1)

plt.figure(figsize=(7, 5))

plt.scatter(
  df_1.iloc[:, 0],
  df_1.iloc[:, 1], 
  c=y_pred
)

plt.title('Clustering')

plt.xlabel('height')
plt.ylabel('weight')

plt.show()

km.predict(df_1)

print(df_1)

df_1['cluster_id'] = y_pred
km.cluster_centers_, km.labels_
```
