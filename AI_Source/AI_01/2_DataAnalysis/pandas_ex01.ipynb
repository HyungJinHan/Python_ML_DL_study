```python
s = pd.Series([9904312, 3448737, 2890451, 2466052],
             index=["서울", "부산", "인천", "대구"])

s2 = pd.Series({'서울':9631482, '부산':3393191, '인천':2632035, '대전':1490158},
              index=['부산', '서울', '대전', '인천'])

ds = s - s2
ds
# 동일한 인덱스 이름을 찾아서 뺄셈 수행
```

```python
s.values - s2.values
# 인덱스의 순서대로 뺄셈 수행
```

```python
ds.notnull()
# 값이 없으면 False, 값이 있으면 True 출력
```

```python
ds[ds.notnull()]
# True에 해당하는 값만 출력
```

```python
rs = (s - s2) / s2 * 100
# 인구 증가율 구하기

rs = rs[rs.notnull()]
rs
```

```python
rs['부산'] = 2.0
# rs.부산 = 2.0
# 수정
rs['대구'] = 1.5
# rs.대구 = 1.5
# 추가
del rs['서울']
# del rs.서울
# 삭제
rs['서울'] = 2.8
# rs.서울 = 2.8
# 추가
rs
```
