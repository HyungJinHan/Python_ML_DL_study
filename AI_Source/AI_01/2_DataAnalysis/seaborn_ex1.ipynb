```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus'] = False # -기호 나오게 표시해줌

titanic = sns.load_dataset('titanic') #타이타닉호 데이터
titanic
```

```python
iris = sns.load_dataset('iris') #붓꽃 데이터(분류용)
iris
```

```python
iris.info()
```

```python
iris.value_counts('species')
```

```python
x = iris.petal_length.values

plt.rcParams["font.family"] = 'Malgun Gothic'
# 한글 처리
sns.rugplot(x)

plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Rug Plot")
plt.show()
```

```python
x = iris.petal_length.values

sns.kdeplot(x)

plt.rcParams["font.family"] = 'Malgun Gothic'
# 한글 처리

plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Kernel Density Plot")
plt.show()
```

```python
x = iris.petal_length.values

sns.displot(x, kde=True, rug=True)

plt.rcParams["font.family"] = 'Malgun Gothic'
# 한글 처리

plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Dist Plot")
plt.show()
```

```python
x = iris.petal_length.values

sns.distplot(x, kde=True, rug=True)

plt.rcParams["font.family"] = 'Malgun Gothic'
# 한글 처리

plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Dist Plot")
plt.show()
```

```python
sns.jointplot(x = 'sepal_length', y = 'sepal_width', data = iris, kind = 'scatter')

plt.rcParams['font.family'] = 'Malgun Gothic' # 한글 처리
plt.suptitle('꽃받침의 길이와 넓이의 Joint Plot', y = 1.02)
plt.show()
```

```python
sns.jointplot(x = 'sepal_length', y = 'sepal_width', data = iris, kind = 'kde', color = 'k')

plt.rcParams['font.family'] = 'Malgun Gothic' # 한글 처리
plt.suptitle('꽃받침의 길이와 넓이의 Joint Plot', y = 1.02)
plt.show()
```

```python
sns.jointplot(x = 'sepal_length', y = 'sepal_width', data = iris, kind = 'hex', color = 'r')

plt.rcParams['font.family'] = 'Malgun Gothic' # 한글 처리
plt.suptitle('꽃받침의 길이와 넓이의 Joint Plot', y = 1.02)
plt.show()
```

```python
sns.pairplot(iris)
plt.title('Iris Data의 Pair Plot')
plt.show()
```

```python
sns.pairplot(iris, hue = 'species', markers = ['o', 's', 'D'])
plt.title('Iris Data의 Pair Plot, Hue로 꽃의 종을 시각화', x = -1.3, y = 4.5)
plt.show()
```

```python
titanic = sns.load_dataset('titanic')

titanic_size = titanic.pivot_table(index = 'class',
                                   columns = 'sex',
                                   aggfunc = 'size')

sns.heatmap(titanic_size,
            cmap = sns.light_palette('brown', as_cmap = True),
            annot = True, fmt = 'd')

plt.title('Heatmap')
plt.show()
```

```python
flights = sns.load_dataset('flights')

flights_passengers = flights.pivot('month', 'year', 'passengers')

plt.title('연도, 월 별 승객 수에 대한 Heatmap')

sns.heatmap(flights_passengers,
            annot = True, # 수치를 텍스트로 표시
            fmt = 'd', # 수치를 정수형으로 표시
            linewidths = 1)

plt.show()
```

```python
tips = sns.load_dataset('tips')

sns.barplot(x = 'day', y = 'total_bill', data = tips)
plt.title('요일 별, 전체 팁')
plt.show()
```

```python
sns.barplot(x = 'day',
            y = 'total_bill',
            hue = 'sex', # hue 사용
            data = tips)

plt.title('요일 별, 성별 전체 팁의 Histogram')
plt.show()
```

```python
sns.boxplot(x="day",
            y="total_bill",
            data=tips) 

plt.title("요일 별 전체 팁의 Box Plot") 
plt.show()
```

```python
import pandas as pd
df_cases = pd.read_csv("https://api.vitaldb.net/cases")  # clinical information

df_vital = df_cases[['age', 'sex', 'bmi', 'height']].head(51)

sns.boxplot(x = 'age', y = 'height', hue = "sex", data = df_vital)

plt.title('Vital DB')
plt.show()

df_vital
```

```python
tips = sns.load_dataset('tips')

sns.violinplot(x = 'day',
               y = 'total_bill',
               data = tips)

plt.title('요일 별 전체 팁의 Violin Plot')
plt.show()
```

```python
tips = sns.load_dataset('tips')

sns.stripplot(x = 'day',
              y = 'total_bill',
              data = tips,
              jitter = True)
```

```python
tips = sns.load_dataset('tips')

sns.swarmplot(x = 'day',
              y = 'total_bill',
              color = 'green',
              data = tips)

plt.title('요일 별 전체 팁의 Swarm Plot')
plt.show()
```

```python
tips = sns.load_dataset('tips')

sns.swarmplot(x = 'day',
              y ='total_bill',
              palette = 'dark:gray',
              hue = 'sex',
              data = tips)

plt.title('요일 별 전체 팁의 Swarm Plot')
plt.show()
```

```python
data = titanic[titanic.survived.notnull()]

sns.catplot(x = 'age',
            y = 'sex',
            hue = 'survived',
            row = 'class',
            data = data,
            kind = 'violin',
            split = True,
            height = 2,
            aspect = 4)

plt.title('각 클래스의 성별 생존자 수의 Catplot')
plt.show()
```

```python
import numpy as np
tips = sns.load_dataset('tips')

sns.boxplot(x = 'tip',
            y = 'day',
            data = tips,
            whis = np.inf)

sns.stripplot(x = 'tip',
              y = 'day',
              data = tips,
              jitter = True,
              color = 'cyan')

plt.title('Boxplot과 Strip Plot로 표현한, 요일 별 팁')
plt.show()
```
