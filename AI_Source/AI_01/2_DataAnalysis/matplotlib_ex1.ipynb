```python
x = np.arange(-5, 5, 0.1)
y1 = x ** 2 - 2
y2 = 20 * np.cos(x) ** 2
```

```python
import matplotlib as mpl

mpl.rcParams['axes.unicode_minus'] = False
# - 표시 유니코드 설정

X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title('LaTeX, 문자열로 tick label 정의')
plt.plot(X, C)

plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
            [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
# 틱 라벨 문자열에는 $$ 사이에 LaTeX 수학 문자 식을 리스트로 넣어 지정 가능

plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])

plt.grid(True)# 격자 표시 (기본 값 : False)

plt.show()
```

```python
x1 = np.linspace(0.0, 5.0, num=50)
x2 = np.linspace(0.0, 2.0, num=50)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.suptitle('=== Main Title ===')

plt.subplot(2, 1, 1)
# = plt.subplot(211)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplot')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)
# = plt.subplot(212)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
# plot 간의 간격을 자동 조정해준다.
plt.show()
```

```python
fig, ax0 = plt.subplots() # 복수 개의 subplot생성

ax1 = ax0.twinx() # twinx() : x축을 공유하는 새로운 Axes객체 생성

ax0.set_title('2개의 y축을 한 figure에서 사용하기')

ax0.plot([10, 5, 2, 9, 7], 'r-', label='y0(Red)')
ax0.set_ylabel('y0(Red)')
ax0.set_xlabel('공유되는 x축')
ax0.grid(False)

ax1.plot([100, 200, 220, 180, 120], 'g--', label='y1(Green)')
ax1.set_ylabel('y1(Green)')
ax1.grid(False)
plt.show()
```

```python
def f(x, y):
    return (1 - x / 2 + x**5 + y**3) * np.exp(-x**2 - y**2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
XX, YY = np.meshgrid(x, y)
ZZ = f(XX, YY)

plt.title('Contour Plots')
plt.contourf(XX, YY, ZZ, alpha=0.75, cmap='twilight') # cmap='hot'
plt.contour(XX, YY, ZZ, colors='black')
plt.show()
```

```python
y1 = [70, 80, 50]
y2 = [30, 60, 90]

x = np.arange(len(y1))
err1 = np.abs(np.random.randn(3)) + 5 # 임의로 지정
err2 = np.abs(np.random.randn(3)) + 5 # 임의로 지정

xlabel = ['국어', '영어', '수학']

plt.title('Bar Chart')

plt.bar(x, y1,
        yerr = err1,
        # 표준 편차
        ecolor = 'yellow',
        width = 0.3,
        color = 'red',
        align='center',
        label = '반 평균')

plt.bar(x+0.4, y2,
        yerr=err2,
        width=0.3,
        color='blue',
        align='center',
        label="반 평균")

plt.xticks(x + 0.4 / 2, xlabel)
# + 0.4 : 차트끼리 간격 띄우기
plt.ylim(0, 100)
plt.xlabel('과목')
plt.ylabel('점수')
plt.legend()
plt.show()
```

```python
y1 = [70, 90, 80]
y2 = [60, 70, 90]
x = np.arange(len(y1))

err1 = np.abs(np.random.randn(3)) + 5 # 임의로 지정
err2 = np.abs(np.random.randn(3)) + 5 # 임의로 지정

xlabel = ['국어', '영어', '수학']

plt.title("Barh Chart")
plt.barh(x, y1,
         height=0.3,
         xerr=err1,
         ecolor='yellow',
         color='red',
         align='center',
         label="점수")

plt.barh(x+0.3,
         y2,
         height=0.3,
         xerr=err2, color='blue', align='center',
         label="반평균")

plt.yticks(x+0.3/2, xlabel)
plt.xlim(0, 100)
plt.xlabel("점수")
plt.ylabel("과목")

plt.legend()
plt.show()
```

```python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15.55, 30, 45, 9.45]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)

plt.title("Pie Chart")
plt.pie(sizes,
        explode=explode,
        labels=labels,
        colors=colors,
        autopct='%1.2f%%',
        shadow=True,
        startangle=90)
plt.axis('equal')
plt.show()
```
