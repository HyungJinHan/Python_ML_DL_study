```python
# 범주형 변수까지 모든 데이터 다 넣음
import pandas as pd

df = pd.read_csv('C:/Users/USER/Desktop/regression220926/train.csv')
df
```

```python
df.isnull().sum()
# ID, ADDRESS 컬럼 삭제
# NEAREST_SCH_RANK는 전체 자료의 30% 이상이 결측치이므로 삭제
```

```python
df.drop(['ID','ADDRESS','NEAREST_SCH_RANK'],axis=1, inplace=True)
df.head()
```

```python
# 결측치
# GARAGE(차고 수)는 평균으로, BUILD_YEAR(건축년도)는 제거
df['GARAGE'].fillna(df['GARAGE'].mean(), inplace=True)
```

```python
df.dropna(inplace=True)
df.shape
```

```python
df.isnull().sum()
```

```python
# suburb(동네 이름), nearest_stn(근처 역 정보), date_sold(판매된 날짜), nearest_sch(근교 학교)
# 위의 컬럼은 카테고리 변수

#df.corr()
# 근처 역까지 거리, 근처 학교까지 거리
# 건물면적-침실 개수, 건물면적-가격
# 위의 3가지가 상대적으로 상관관계 높음

# 가격과의 상관관계만 보면 상관관계가 높은 feature가 없기 때문에 feature간의 관계도 고려해야 할 듯
```

```python
df['SUBURB']=df['SUBURB'].astype('category')
df['NEAREST_STN']=df['NEAREST_STN'].astype('category')
df['DATE_SOLD']=df['DATE_SOLD'].astype('category')
df['NEAREST_SCH']=df['NEAREST_SCH'].astype('category')
df['BUILD_YEAR']=df['BUILD_YEAR'].astype('int64')
df.info()
```

```python
import pandas as pd

df = pd.get_dummies(df, columns=['SUBURB','NEAREST_STN','DATE_SOLD','NEAREST_SCH'])
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df.iloc[:,1:])
```

```python
df.columns
```

```python
col = []
for i in range(0,len(df.columns)):
    col.append(df.columns[i].replace(" ","_").replace("-","_").replace("\r","").replace("'",""))

df.columns=col
df.head()
```

```python
import statsmodels.formula.api as smf

formula = 'PRICE~'+"+".join(df.iloc[:,1:].columns)
model = smf.ols(formula=formula, data=df).fit()
model.summary()

# 결정계수 0.814
```
