```python
# 모르겠어서 문자 다 빼고 숫자로만 해보려고 다시 불러옴
import pandas as pd

df2 = pd.read_csv('C:/Users/USER/Desktop/regression220926/train.csv')
df2.head()
```

```python
df2.drop(['ID','ADDRESS','SUBURB','NEAREST_STN','DATE_SOLD','NEAREST_SCH'],axis=1,inplace=True)
df2.head()
```

```python
df2.drop(['NEAREST_SCH_RANK'],axis=1,inplace=True)
df2['GARAGE'].fillna(df2['GARAGE'].mean(), inplace=True)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(df2.iloc[:,1:],df2['PRICE'], random_state=5)
```

```python
# minmax scaling
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)
```

```python
# 다중회귀
import statsmodels.formula.api as smf

formula = "PRICE~"+"+".join(df2.iloc[:,1:].columns)

model=smf.ols(formula=formula, data=df2).fit()
model.summary()

# 수치 데이터만 넣었을 때 결정계수는 0.578
```
