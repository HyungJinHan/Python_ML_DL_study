```python
import pandas as pd
import numpy as np
```

```python
home_df = pd.read_csv('./train.csv')
home_test_df = pd.read_csv('./test.csv')
# 타겟 데이터
```

```python
home_df[
  [
    'PRICE',
    'NEAREST_SCH_RANK'
  ]
].groupby(
  [
    'NEAREST_SCH_RANK',
    'PRICE'
  ],
  as_index=False
).mean().sort_values(by='PRICE', ascending=False)
```

```python
home_df.isnull().sum()
```

```python
home_df['GARAGE'].fillna(home_df['GARAGE'].value_counts().idxmax(), inplace=True)

home_df
```

```python
home_df['NEAREST_SCH_RANK'].fillna(home_df['NEAREST_SCH_RANK'].mean(), inplace=True)

home_df
```

```python
home_df['BUILD_YEAR'].fillna(home_df['BUILD_YEAR'].value_counts().idxmax(), inplace=True)

home_df
```

```python
train_df = home_df.drop(['PRICE'], axis=1)
train_df
```

```python
test_df = home_df[['PRICE']]
test_df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train_df, test_df, random_state=14)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
# 학습

score = model.score(X_test, y_test)
# 정확도 체크
score
```
